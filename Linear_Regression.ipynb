```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('AREA')
plt.ylabel('PRICE')
plt.scatter(df.area,df.price,color='blue',marker='*')
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price
```

```python
model = linear_model.LinearRegression()
model.fit(new_df,price)
```

```python
model.predict([[1000]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
new_df2=pd.read_csv('areas.csv')
new_df2.head(3)
```

```python
p = model.predict(new_df2)
p
```

```python
new_df2['prices']=p
new_df2
```

```python
new_df2.to_csv("prediction.csv")
```

```python
%matplotlib inline
plt.xlabel('AREA')
plt.ylabel('PRICE')
plt.scatter(df.area,df.price,color='blue',marker='*')
plt.plot(new_df2.area,new_df2.prices,color='red')
```
