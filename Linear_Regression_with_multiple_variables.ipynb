```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('homeprices_multi.csv')
df
```

```python
df.bedrooms.median()
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
model = linear_model.LinearRegression()
model.fit(df.drop('price',axis = 'columns'),df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[3000,5,3]])
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color = 'red',marker = '+')
plt.plot(df.area,model.predict(df[['area','bedrooms','age']]),color = 'blue')
```

```python
%matplotlib inline
plt.xlabel('bedrooms')
plt.ylabel('price')
plt.scatter(df.bedrooms,df.price,color = 'red',marker = '+')
plt.plot(df.bedrooms,model.predict(df[['area','bedrooms','age']]),color = 'blue')
```

```python
%matplotlib inline
plt.xlabel('age')
plt.ylabel('price')
plt.scatter(df.age,df.price,color = 'red',marker = '+')
plt.plot(df.age,model.predict(df[['area','bedrooms','age']]),color = 'blue')
```
